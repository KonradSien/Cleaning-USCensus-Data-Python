```python
import pandas as pd
import numpy as np
import os
```

```python
path='C:/Users/nexia/Desktop/Jupyter projekty/US-Census-Data'

files=[file for file in os.listdir(path)]

df=pd.DataFrame()

for file in files:
    current_data= pd.read_csv(path+'/'+file)
    df=pd.concat([df,current_data])

df.to_csv('all_us_census_data.csv',index=False)
```

#### check nan

We can fill in those nans by using pandasâ€™ .fillna() function.

You have the TotalPop per state, and you have the Men per state.
As an estimate for the nan values in the Women column, you could use the TotalPop of that state minus the Men for that state.

```python
df.Women=df.Women.fillna(df.TotalPop-df.Men)
df.Women
```

```python
df=df.drop_duplicates()
```

```python
df.Women=df.Women.astype('int64')
```

```python
df.Pacific=df.Pacific.fillna(100-(df.Hispanic+df.White+df.Black+df.Native+df.Asian))
df.Pacific=np.round(df.Pacific,2)
```

```python
print(df.isnull().sum())
```

```python
df.Hispanic=df.Hispanic.str.replace('%','')
df.White=df.White.str.replace('%','')
df.Black=df.Black.str.replace('%','')
df.Native=df.Native.str.replace('%','')
df.Asian=df.Asian.str.replace('%','')
df.Pacific=df.Pacific.str.replace('%','')
```

```python
df.Hispanic=pd.to_numeric(df.Hispanic)
df.White=pd.to_numeric(df.White)
df.Black=pd.to_numeric(df.Black)
df.Native=pd.to_numeric(df.Native)
df.Asian=pd.to_numeric(df.Asian)
df.Pacific=pd.to_numeric(df.Pacific)
```
